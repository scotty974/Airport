```python
import pandas as pd
```

```python
airport_data =pd.read_csv('data/airport.txt')
airport_data.head()
```

```python
flights_data =pd.read_csv('data/flights.csv')
flights_data.head()
```

```python
dest=flights_data[['dest']].drop_duplicates()
faa=airport_data[['faa']].drop_duplicates()
dest
faa
df_all = faa.merge(dest, left_on='faa', right_on='dest', how='right', indicator=True)
df_all
df_all._merge.value_counts()
```
